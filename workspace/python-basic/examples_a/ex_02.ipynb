```python
data = []
with open("data-files/population.csv", encoding="utf-8") as f:
    skip = True
    while (True):
        row = f.readline()
        # if not row:
        if len(row) == 0:
            print("EOF : ", type(row))
            break
        
        # row = row.split("\",\"") # ['a', 'b', 'c']
        row = row.split('","') # ['a', 'b', 'c']
        row2 = []
        # for c in row:
        for idx, c in enumerate(row):
            d = c.replace("\"", "").replace("\n", "").replace(",", "")
            if (skip == False and idx != 0):
                d = int(d)
            row2.append(d)
        data.append(row2)

        skip = False

        
```

```python
import csv # 내장 모듈 csv 가져오기
```

```python
data = [] # 리스트 만들기
with open("data-files/population.csv", encoding="utf-8") as f:
    reader = csv.reader(f) # csv 처리 : ,로 분리하고 "와 같은 데이터 구분 표식 제거하고 리스트로 데이터 반환

    next(reader) # 한 행 읽기 ( 첫 번째 행 skip )
    for _, line in enumerate(reader):
        row = []
        for idx2, c in enumerate(line):
            d = c.replace(",", "")
            if (idx2 != 0):
                d = int(d)
            row.append(d)
        data.append(row)
        
```

```python
# 인구가 가장 많은 행정구역
max_pop = ['', 0]
for row in data:
    if max_pop[1] < row[1]:
        max_pop[0] = row[0]
        max_pop[1] = row[1]
```
